# Query Question I

## Question

- The career summary integrates information from Lahman2017. The relevant tables are:
    - People
    - Appearances
    - Batting
    - Pitching
    - Fielding
    
    
- A sample of running the query is below.


- Columns:
    - _Games_ is $G\_all$ from _Appeareances._
    - Batting: _Hits_ is sum of _H,_ _At Bats_ is sum of _AB,_ _RBIs_ is sum of _RBI._ _Average_ is the sum of _H_ divided by the sum of _AB._
    - Pitching: $P\_wins$ is the sum of _W_ and $P\_loss$ is the sum of _L._
    - Fielding: _Put Outs_ is sum of _PO,_ _Assists_ is sum of _A_ and _Errors_ is some of _E._
    
    
- All numeric types must be integers, except for _Average._ You must round _Average_ to 3 decimal places.


- Your column names must match.


- You should be able to figure out what "--" means and how to produce it.


## Answer

````python
%%sql
use lahman2017;
SELECT B.playerID AS playerID, B.teamID AS teamID, B.yearID AS yearID, CAST(SUM(A.G_all)AS unsigned) AS Games, 
        CAST(B.H AS unsigned) AS Hits, CAST((B.AB) AS unsigned) AS At_Bats, 
        ifnull(ROUND((B.H)/(B.AB),3), '--') AS average, CAST((B.HR) AS unsigned) AS HRs,
        CAST((B.RBI) AS unsigned) AS RBIs,
        CAST(P.W AS unsigned) AS P_wins, CAST(P.L AS unsigned) AS P_loss, 
        CAST((F.PO) AS unsigned) AS Put_Outs, CAST((F.A) AS unsigned) AS Assists, 
        CAST((F.ER) AS unsigned) AS Errors
FROM (SELECT playerID, teamID, yearID, G_all FROM Appearances) AS A
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(H) AS H, SUM(AB) AS AB, SUM(RBI) AS RBI, SUM(HR) AS HR
        FROM Batting
        GROUP BY playerID, teamID, yearID
    ) AS B 
        ON A.playerID = B.playerID AND A.teamid = B.teamID AND A.yearID = B.yearID
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(W) AS W, SUM(L) AS L
        FROM Pitching
        GROUP BY playerID, teamID, yearID
    ) AS P 
        ON A.yearID = P.yearID AND A.teamID = P.teamID AND A.playerID = P.playerID  
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(PO) AS PO, SUM(A) AS A, SUM(E) AS ER
        FROM Fielding 
        GROUP BY playerID, teamID, yearID
    )AS F 
        ON A.playerID = F.playerID AND A.teamID = F.teamID AND A.yearID = F.yearID
GROUP BY A.playerID,A.teamID, A.yearID
LIMIT 20;
````

# Query Question II

## Question

- In the simple version of Moneyball, we tried to predict a team's wins based on a team's performance statistics.


- A simple sample of the initial data we loaded is:

````python
%%sql
use lahman2017;
select teamid, yearid, w, ab, h, `2b` as b2, `3b` as b3, BB, hr, era
    from lahman2017.teams
    where yearid >= 2000
    order by yearid asc
    limit 10;
````

- The Boston Red Sox hit 167 homeruns (hr) in 2000. Was 167 homeruns good or bad? You cannot answer the question in isolation. The answer depends on the homerun production of other teams.


- Normalizing performance require two steps:
    1. Computing average and aggregate behavior in a season over all teams.
    2. Understanding how a team's metrics compared to aggregates and averages for the season.
    
    
- Write __a single query__ that produces a relative performance table per team, per year.
    - For w, ab, h, b2, b3, hr, and bb, the corresponding column is the team's value divided by the total over all teams for the year.
    - For ERA, the value is the team's ERA relative to the average ERA over all teams for the year.

## Answer

````python
%%sql
WITH tmp(yearid, total_w, total_ab, total_h, total_b2, total_b3, total_hr, total_bb, avg_era) AS
    (SELECT yearid, sum(w) as total_w, sum(ab) as total_ab, sum(h) as total_h, sum(2b) as total_b2, sum(3b) as total_b3,
            sum(hr) as total_hr, sum(bb) as total_bb, avg(era) as avg_era
    FROM Teams
    GROUP BY yearID
    HAVING yearID >=2000)
SELECT t.teamID, t.yearID, round(t.w/total_w, 3) as w, round(t.ab/total_ab, 3) as ab,
    round(t.h/total_h, 3) as h, round(t.2b/total_b2, 3) as b2, round(t.3b/total_b3, 3) as b3,
    round(t.hr/total_hr, 3) as hr, round(t.bb/total_bb, 3) as bb, round(t.era/avg_era,3) as rel_era
FROM Teams as t
JOIN tmp as p
ON t.yearID = p.yearID
WHERE t.yearid >= 2000
LIMIT 20;
````

````python
%%sql
use lahman2017;
SELECT B.playerID AS playerID, B.teamID AS teamID, B.yearID AS yearID, CAST(SUM(A.G_all)AS unsigned) AS Games, 
        CAST(B.H AS unsigned) AS Hits, CAST((B.AB) AS unsigned) AS At_Bats, 
         ifnull(ROUND((B.H)/(B.AB),3), '--') AS average, CAST((B.HR) AS unsigned) AS HRs,
        CAST((B.RBI) AS unsigned) AS RBIs,
        CAST(P.W AS unsigned) AS P_wins, CAST(P.L AS unsigned) AS P_loss, 
        CAST((F.PO) AS unsigned) AS Put_Outs, CAST((F.A) AS unsigned) AS Assists, 
        CAST((F.ER) AS unsigned) AS Errors
FROM (SELECT playerID, teamID, yearID, G_all FROM Appearances) AS A
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(H) AS H, SUM(AB) AS AB, SUM(RBI) AS RBI, SUM(HR) AS HR
        FROM Batting
        GROUP BY playerID, teamID, yearID
    ) AS B 
        ON A.playerID = B.playerID AND A.teamid = B.teamID AND A.yearID = B.yearID
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(W) AS W, SUM(L) AS L
        FROM Pitching
        GROUP BY playerID, teamID, yearID
    ) AS P 
        ON A.yearID = P.yearID AND A.teamID = P.teamID AND A.playerID = P.playerID  
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(PO) AS PO, SUM(A) AS A, SUM(E) AS ER
        FROM Fielding 
        GROUP BY playerID, teamID, yearID
    )AS F 
        ON A.playerID = F.playerID AND A.teamID = F.teamID AND A.yearID = F.yearID
GROUP BY A.playerID,A.teamID, A.yearID
HAVING playerid = 'willite01';
````

````python
%%sql
use lahman2017;
CREATE TABLE `player_performance_summary` AS 
SELECT B.playerID AS playerID, B.teamID AS teamID, B.yearID AS yearID, CAST(SUM(A.G_all)AS unsigned) AS Games, 
        CAST(B.H AS unsigned) AS Hits, CAST((B.AB) AS unsigned) AS At_Bats, 
        ROUND((B.H)/nullif(B.AB, 0),3) AS average, CAST((B.HR) AS unsigned) AS HRs,
        CAST((B.RBI) AS unsigned) AS RBIs,
        CAST(P.W AS unsigned) AS P_wins, CAST(P.L AS unsigned) AS P_loss, 
        CAST((F.PO) AS unsigned) AS Put_Outs, CAST((F.A) AS unsigned) AS Assists, 
        CAST((F.ER) AS unsigned) AS Errors
FROM (SELECT playerID, teamID, yearID, G_all FROM Appearances) AS A
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(H) AS H, SUM(AB) AS AB, SUM(RBI) AS RBI, SUM(HR) AS HR
        FROM Batting
        GROUP BY playerID, teamID, yearID
    ) AS B 
        ON A.playerID = B.playerID AND A.teamid = B.teamID AND A.yearID = B.yearID
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(W) AS W, SUM(L) AS L
        FROM Pitching
        GROUP BY playerID, teamID, yearID
    ) AS P 
        ON A.yearID = P.yearID AND A.teamID = P.teamID AND A.playerID = P.playerID  
    LEFT JOIN (
        SELECT playerID, teamID, yearID, SUM(PO) AS PO, SUM(A) AS A, SUM(E) AS ER
        FROM Fielding 
        GROUP BY playerID, teamID, yearID
    )AS F 
        ON A.playerID = F.playerID AND A.teamID = F.teamID AND A.yearID = F.yearID
GROUP BY A.playerID,A.teamID, A.yearID;
````

````python
%%sql
use rs3878_s2019_w4111_final;
drop table if exists player_career_performance_summary;
create table `player_career_performance_summary` as
select playerid, round(avg(games),3) as avg_games, round(avg(hits),3) as avg_hits, 
    round(avg(at_bats),3) as avg_at_bats, round(avg(average),3) as average, 
    round(avg(HRs),3) as avg_HRs, round(avg(rbis),3) as avg_rbis,
    round(avg(P_wins),3) as avg_p_wins, round(avg(p_loss),3) as avg_p_loss,
    round(avg(Put_outs),3) as avg_put_outs, round(avg(assists),3) as avg_assists,
    round(avg(errors),3) as avg_errors
from player_performance_summary
group by playerid
having sum(games)>=100;
````

#### at most 25 people in the team

````python
%%sql
create definer=`root`@`localhost` 
TRIGGER `rs3878_s2019_w4111_final`.`fantasy_team_BEFORE_INSERT` BEFORE INSERT ON `fantasy_team_players` FOR EACH ROW
BEGIN
    declare u varchar(32);
    declare c int;
    SELECT count(*) INTO c FROM fantasy_team_players;
    set u = (select user()); 
    if u != 'root@localhost' then
        if c > 25 then
            signal sqlstate '45002'
                set message_text="too many players";
        end if;
    end if;
END
````

````python
%%sql
create definer=`root`@`localhost` 
TRIGGER `rs3878_s2019_w4111_final`.`fantasy_team_atleast10_BEFORE_INSERT` BEFORE INSERT ON `fantasy_team_players` FOR EACH ROW
BEGIN
    declare u varchar(32);
    declare c int;
    SELECT count(*) INTO c FROM fantasy_team_players join player_career_performance_summary on playerid where avg_p_wins = null;
    set u = (select user()); 
    if u != 'root@localhost' then
        if c > 10 then
            signal sqlstate '45002'
                set message_text=" at least 10 players with non-null value for p_wins";
        end if;
    end if;
END
````
